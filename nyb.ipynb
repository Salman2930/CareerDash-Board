```python
import requests
from bs4 import BeautifulSoup as sk
import re
import pandas as pd
import json

url = 'https://api.nbyula.com/discovery/v1/jobs'

r = requests.get(url)
soup = r.json()
final_data = []
for berry in soup:
    content = {}
    content['Job Title'] = berry['title']
    content['Job Function'] = berry['jobFunction']
    content['Experiance'] = berry['experienceLevel']
    content['Location'] = berry['location']
    content['Apply Here'] = 'https://nbyula.com/job/' + berry['_id']
    final_data.append(content)
nby_df_data = pd.DataFrame(final_data)
nby_df_data
```
